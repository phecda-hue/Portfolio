```python
# 예제 4
import pandas as pd
df=pd.read_csv("data.csv")
print(df.isnull().sum()) # .isnull()은 데이터 프레임에서 결측치가 있는 값을 T/F로 알려줌(True가 결측치)
# .sum()을 붙이면 데이터 프레임의 각 열에 몇 개의 결측치가 있는지 알려줌
df_filled=df.fillna(0) # .fillna(숫자)는 데이터 프레임에서 결측치를 입력한 숫자로 대체함
print(df_filled)
```

연습문제 2. 데이터프레임에서 여러 칼럼 선택하기

df[["column1", "column2"]]를 사용하여 여러 개의 컬럼을 선택하라.

```python
import pandas as pd
df=pd.read_csv("data.csv")
df_name_price=df[["name", "price"]] # 열 선택 시 대괄호를 두 개 쳐야 함
# 데이터프레임[["이름1", "이름2"]]는 데이터 프레임에서 이름1, 이름2의 열만을 선택함
print(df_name_price)
```

연습문제 3. 데이터프레임의 특정 행 선택하기

.iloc[] 또는 .loc[]을 사용하여 특정 행을 선택하는 방법을 연습하라.

```python
import pandas as pd
df=pd.read_csv("data.csv")
df_012=df.iloc[[0, 1, 2], ] # .iloc[[n], [m]]은 입력한 번호의 행과 열만을 지정함
print(df_012)
# .loc[[행], [열]]은 모든 행의 이름이 지정되었을 경우 입력한 이름의 행과 열만을 지정함
```

연습문제 4. 결측치를 평균으로 채우기

.fillna(df["컬럼명"].mean())을 사용하여 결측치를 평균값으로 채워라.

```python
import pandas as pd
df=pd.read_csv("data.csv")
print(df.isnull().sum())
df["price"]=df["price"].fillna(df["price"].mean()) # .mean()은 해당 열의 평균값을 구함
# price 열만의 결측치를 대체함
print(df)
```

연습문제 7. 데이터프레임의 중복 행 제거하기

.drop_duplicates()를 사용하여 중복 데이터를 제거하는 방법을 연습하라.

```python
import pandas as pd
df=pd.read_csv("data.csv")
row_to_insert = df.iloc[9].copy() # 10번째 행을 복사한 행을 만듦
# .copy()는 해당 행과 같은 행을 만듦
df_duplicate=pd.concat([df.iloc[:14], pd.DataFrame([row_to_insert]), df.iloc[14:]], ignore_index=True)
print(df_duplicate)
df_delduplicate=df.drop_duplicates() # .drop_duplicates()는 데이터 프레임에서 중복된 행을 삭제함
print(df_delduplicate)
```

연습문제 8. 데이터프레임의 특정 컬럼 삭제하기

.drop(columns=["컬럼명"])을 사용하여 특정 컬럼을 제거하는 방법을 연습하라.

```python
import pandas as pd
df=pd.read_csv("data.csv")
df_id=df.drop(columns=["id"]) #.drop(columns=[열 이름])을 통해 원하는 열을 삭제할 수 있음
print(df_id)
```

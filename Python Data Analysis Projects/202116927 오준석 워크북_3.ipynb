### **ì˜ˆì œ 1: ê²°ì¸¡ì¹˜ ê°œìˆ˜ í™•ì¸ ë° ì²˜ë¦¬**

ğŸ“Œ **ë¬¸ì œ:** `titanic` ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì™€ì„œ, ê° ì»¬ëŸ¼ë³„ ê²°ì¸¡ì¹˜ ê°œìˆ˜ë¥¼ í™•ì¸í•˜ê³ , `age` ì»¬ëŸ¼ì˜ ê²°ì¸¡ì¹˜ë¥¼ í‰ê· ê°’ìœ¼ë¡œ ì±„ìš°ì‹œì˜¤.

ğŸ’¡ **íŒíŠ¸:** `.isnull().sum()`ì„ ì‚¬ìš©í•˜ë©´ ì»¬ëŸ¼ë³„ ê²°ì¸¡ì¹˜ ê°œìˆ˜ë¥¼ í™•ì¸í•  ìˆ˜ ìˆìŒ.

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

print(titanic.isnull().sum()) # .isnull() í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ë©´ ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” ê³³ì˜ ê°’ì´ trueë¡œ ë°”ë€œ
# .sum() í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ë©´ ê°’ì´ trueì¸ ê²ƒì˜ ê°œìˆ˜ë¥¼ ì¶œë ¥

titanic["age"]=titanic["age"].fillna(titanic["age"].mean())
# .fillna() í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ë©´ ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” ë¶€ë¶„ë§Œ íŠ¹ì •í•œ ê°’ìœ¼ë¡œ ì±„ìš¸ ìˆ˜ ìˆ

print(titanic.isnull().sum())
```

### **ì˜ˆì œ 2: ì´ìƒì¹˜ íƒìƒ‰ ë° ì²˜ë¦¬ (IQR í™œìš©)**

ğŸ“Œ **ë¬¸ì œ:** `titanic` ë°ì´í„°ì—ì„œ `fare` ì»¬ëŸ¼ì˜ ì´ìƒì¹˜ë¥¼ IQR ë°©ë²•ì„ ì‚¬ìš©í•˜ì—¬ íƒìƒ‰í•˜ì‹œì˜¤.

ğŸ’¡ **íŒíŠ¸:** ì´ìƒì¹˜ëŠ” **IQR(Interquartile Range)**ì„ ì´ìš©í•˜ì—¬ íƒìƒ‰í•  ìˆ˜ ìˆìŒ.

```python
import numpy as np
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
Q1 = df["fare"].quantile(0.25)
Q3 = df["fare"].quantile(0.75) # .quantile(0.25 / 0.5 / 0.75) í•¨ìˆ˜ë¥¼ í†µí•´ ì œ 1, 2, 3 ì‚¬ë¶„ìœ„ìˆ˜ë¥¼ êµ¬í•  ìˆ˜ ìˆìŒ
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# ì´ìƒì¹˜ëŠ” ì œ1 ì‚¬ë¶„ìœ„ìˆ˜ë³´ë‹¤ IQRì˜ 1.5 ì´ìƒ ë°° ì‘ì€ ê°’ ë˜ëŠ” ì œ3 ì‚¬ë¶„ìœ„ìˆ˜ë³´ë‹¤ IQRì˜ 1.5ë°° ì´ìƒ í° ê°’

outliars = df[(df["fare"] < lower_bound) | (df["fare"] > upper_bound)]
# ë°ì´í„°í”„ë ˆì„[ì¡°ê±´1 | ì¡°ê±´2]ì˜ í˜•ì‹ì„ ì‚¬ìš©í•˜ë©´ ë‘ ê°€ì§€ ì¡°ê±´ ì¤‘ í•˜ë‚˜ë¥¼ ë§Œì¡±í•˜ëŠ” í–‰ì„ ì„ íƒí•  ìˆ˜ ìˆìŒ
print(outliars)
```

ğŸ’¡ **íŒíŠ¸:** `pd.get_dummies(df["sex"])`ë¥¼ ì‚¬ìš©í•˜ë©´ ì›-í•« ì¸ì½”ë”©ì„ ì ìš©í•  ìˆ˜ ìˆìŒ.

### **ì˜ˆì œ 4: Min-Max ìŠ¤ì¼€ì¼ë§ ì ìš©**

ğŸ“Œ **ë¬¸ì œ:** `titanic` ë°ì´í„°ì—ì„œ `fare` ì»¬ëŸ¼ì„ 0ê³¼ 1 ì‚¬ì´ ê°’ìœ¼ë¡œ ë³€í™˜í•˜ì‹œì˜¤.

ğŸ’¡ **íŒíŠ¸:** `MinMaxScaler()`ë¥¼ ì‚¬ìš©í•˜ë©´ ê°’ì„ 0~1 ì‚¬ì´ë¡œ ì¡°ì •í•  ìˆ˜ ìˆìŒ.

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

df = sns.load_dataset("titanic")
scaler = MinMaxScaler()
df["fare_scaled"] = scaler.fit_transform(df[["fare"]])
# scaler.fit_transform(ë°ì´í„°í”„ë ˆì„[["ì—´"]]) í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ë©´ í•´ë‹¹ ì—´ì˜ ê°’ì„ 0~1 ì‚¬ì´ì˜ ê°’ì„ ê°–ëŠ” ê°’ìœ¼ë¡œ ì¹˜í™˜í•  ìˆ˜ ìˆìŒ
print(df[["fare", "fare_scaled"]].head())
```

### **ì˜ˆì œ 5: ìƒˆë¡œìš´ í”¼ì²˜ ìƒì„±í•˜ê¸° (ì´ ì§€ë¶ˆ ê¸ˆì•¡ ìƒì„±)**

ğŸ“Œ **ë¬¸ì œ:** `fare`ì™€ `tip`ì˜ í•©ì„ ìƒˆë¡œìš´ `total_payment` ì»¬ëŸ¼ìœ¼ë¡œ ì¶”ê°€í•˜ì‹œì˜¤.

ğŸ’¡ **íŒíŠ¸:** ìƒˆë¡œìš´ ì»¬ëŸ¼ì„ ìƒì„±í•  ë•Œ `df["ìƒˆë¡œìš´ ì»¬ëŸ¼"] = df["ì»¬ëŸ¼1"] + df["ì»¬ëŸ¼2"]` í˜•ì‹ì„ ì‚¬ìš©í•˜ë©´ ë¨.

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")
df["total_payment"] = df["total_bill"] + df["tip"]
# ë°ì´í„°í”„ë ˆì„["ìƒˆë¡œìš´ ì—´"] = ë°ì´í„°í”„ë ˆì„["ì—´1"] + ë°ì´í„°í”„ë ˆì„["ì—´2"] í˜¹ì€ ë‹¤ë¥¸ ì—°ì‚°ìë¥¼ ì‚¬ìš©í•˜ë©´ ìƒˆë¡œìš´ ì—´ì„ ë§Œë“¤ ìˆ˜ ìˆìŒ
df["ratio"] = df["tip"] / df["total_bill"]

print(df[["total_bill", "tip", "total_payment", "ratio"]])
# ì—¬ëŸ¬ ê°œì˜ ì—´ì„ ì„ íƒí•  ë•ŒëŠ” ëŒ€ê´„í˜¸ë¥¼ ë‘ ê°œ ì‚¬ìš©í•´ì•¼ í•¨
```

1. **ê²°ì¸¡ì¹˜ë¥¼ ì¤‘ì•™ê°’(median)ìœ¼ë¡œ ì±„ìš°ê¸°**
